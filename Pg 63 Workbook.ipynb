```python
import pandas as pd
Location = "datasets/gradedata.csv"
df = pd.read_csv(Location)
df.head()
```

```python
df['gender_new'] = df['gender'].map({'male': 0,'female': 1})
df
```

```python
import statsmodels.formula.api as sm
result = sm.ols(
 formula='grade ~ age + exercise + hours + gender_new',
 data=df).fit()
result.summary()
```
